```python
#Importing libraries
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#loading the data
books_data = pd.read_csv('books.csv')
books_data['index'] = range(1, len(books_data) + 1)
books_data.head()
```

```python
books_data.shape
```

```python
books_data.isnull().sum()
```

```python
#Selecting the relevant features for recommendation system
features = ['title', 'subtitle', 'authors', 'categories', 'average_rating']
features
```

```python
for i in range(1, len(features) - 1):
    books_data[features[i]] = books_data[features[i]].fillna('')
books_data.isnull().sum()
```

```python
books_data['average_rating'] = books_data['average_rating'].fillna(books_data['average_rating'].median())
books_data.isnull().sum()
```

```python
#Combining all the five selected features
combined_features = books_data['title'] + ' ' + books_data['subtitle'] + ' ' + books_data['authors'] + ' ' + books_data['categories'] + ' ' + str(books_data['average_rating'])
combined_features
```

```python
#Converting the text data to feature vectors
vectorizer = TfidfVectorizer()
feature_vector = vectorizer.fit_transform(combined_features)
feature_vector
```

#### Cosine Similarity

```python
# Getting a similary score using cosine similarity
similarity = cosine_similarity(feature_vector)
similarity
```

```python
similarity.shape
```

```python
#Getting the book name
book_name = input("Enter the book name:")
```

```python
#Creating a list of all the bookss given in the dataset
list_titles = books_data['title'].tolist()
list_titles
```

```python
#Finding the close match for the book name given by the user
find_close_match = difflib.get_close_matches(book_name, list_titles)
find_close_match
```

```python
close_match = find_close_match[0]
close_match
```

```python
#Finding the index of the movie with title
index_of_the_book = books_data[books_data.title == close_match]['index'].values[0]
index_of_the_book
```

```python
# getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_book]))
similarity_score
```

```python
len(similarity_score)
```

```python
# sorting the bookss based on their similarity score
sorted_similar_books = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
sorted_similar_books
```

```python
# print the name of similar movies based on the index

print('Books suggested for you : \n')

i = 1

for book in sorted_similar_books:
  index = book[0]
  title_from_index = books_data[books_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
